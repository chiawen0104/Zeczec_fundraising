## Q01.1 Tokenization by jeiba
Generate a new column named "tokenized" to the df

````python
!pip install hanziconv
from hanziconv import HanziConv
````

````python
import jieba
# YOUR CODE SHOULD BE HERE
token = []

for i in df['textOriginal']:
  #print([i])
  tra = HanziConv.toTraditional([i])
  sim = HanziConv.toSimplified(tra)
  sim_cutted = list(jieba.cut(sim))
  #print(sim_cutted)
  #print('-'*50)
  token.append(sim_cutted)

df['tokenized'] = token
df
````

### Q01.5 å¯«ä¸€å‡½å¼å»é™¤æ‰€æœ‰ä¸­æ–‡åœç”¨è©`removeChiStopWords()`
```
def removeChiStopWords(...):
    ...

```

````python
def removeChiStopWords(dict):
    words_clean = {}
    for k,v in dict.items():
        if k not in stopwords:
            words_clean[k] = v
    return words_clean

filtered = removeChiStopWords(filtered)
print(len(filtered))
filtered
````

### Q01.6 å¯«ä¸€å‡½å¼å»é™¤æ‰€æœ‰çš„è‹±æ–‡å­—æ¯èˆ‡æ•¸å­—`removeAlphaNum()`ã€‚
```
def removeAlphaNum(...):
    ...
filtered = removeAlphaNum(filtered)
print(filtered)
```

````python
# YOUR CODE SHOULD BE HERE
# YOU CAN CHANGE THE VARIABLE NAME
num = '0123456789'

def removeAlphaNum(dict):
    w_clean = {}
    for k,v in dict.items():
        flag = 0
        for ch in k: 
          if ch in num:
            flag = 1
            break
        if k.encode().isalpha() == True:
          continue
        if flag == 0:
          w_clean[k] = v
    return w_clean

filtered = removeAlphaNum(filtered)
filtered
````

### Q01.7 å¯«ä¸€å‡½å¼å»é™¤å…¨åŠå½¢æ¨™é»ç¬¦è™Ÿæˆ–ç©ºç™½ï¼š`removalPuncSpace()`ã€‚
```
def removalPuncSpace(...):
    ...
filtered = removalPuncSpaces(filtered)
print(filtered)
```

````python
import unicodedata
def removalPuncSpace(dict):
    wd_clean = {}
    for k,v in dict.items():
        for ch in k:
          if unicodedata.category(ch).startswith('L'):
            #print(k, unicodedata.category(ch))
            wd_clean[k] = v
    return wd_clean

filtered = removalPuncSpace(filtered)
filtered
````
